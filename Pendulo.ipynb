```python
def osc_armonico(y,t,w02 = 1.):
    return array([y[1],-w02*sin(y[0])+0.1*sin(t)])
```

```python
for i in range(3):
    y0,u0 = pi*(2*rand()-1),0 #pi*(2*rand()-1)
    t,solucion = integradorRK4orden(osc_armonico,[y0,u0],[0,500.])
    plot(solucion[:,0],solucion[:,1])
zoom = pi
axes().set_aspect("equal")
grid(True)
xlim(-zoom,zoom)
ylim(-zoom,zoom)
show()
```

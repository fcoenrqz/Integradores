```python
from pylab import *
%matplotlib inline
```

```python
def euler(funcion,p_ini,t_final,h=0.01):
    tiempos = arange(p_ini[0],t_final+h,h)
    y = zeros(len(tiempos))
    y[0] = p_ini[1]
    for i in range(len(tiempos)-1):
        y[i+1] = y[i] + funcion(y[i],tiempos[i])*h
    solucion = [tiempos,y]
    return solucion
```

Ecuaci√≥n diferencial a resolver: 
$$
f'(x) = f(x)
$$

```python
def mi_ed(y,t):
    return y
```

```python
sol = euler(mi_ed,[0,1],3,0.001)
plot(sol[0],sol[1],sol[0],exp(sol[0]))
show()
```

Ecuaci√≥n diferencial a resolver: 
$$
f'(x) =-xf(x)
$$

```python
def mi_ed2(y,t):
    return -t*y
```

```python
sol2 = euler(mi_ed2,[0,1],3,0.1)
plot(sol2[0],sol2[1],sol2[0],exp(-((sol2[0])**2)/2))
show()
```

```python
def mi_ed3(y,t):
    return ((1./(sin(t)+2))*(((2*y/sqrt(3))**2) - 1))
```

```python
hs = [0.5,0.1,0.01,0.001,1e-3]
for h in hs:
    sol3 = euler(mi_ed3,[1,1],3.,h)
    plot(sol3[0],sol3[1],label= "h="+str(h))
    legend(loc = 0)
show()
```

```python
def segundo_orden(y,t):
    return array([y[1],-y[0]])
```

```python
def eulervect(funcion,p_ini,t_final,h=0.01):
    tiempos = arange(p_ini[0],t_final+h,h)
    y = [0]*(len(tiempos))
    y[0] = p_ini[1]
    for i in range(len(tiempos)-1):
        y[i+1] = y[i] + funcion(y[i],tiempos[i])*h
    solucion = [tiempos,y]
    return solucion
```

```python
sol4 = eulervect(segundo_orden,[0,array([0,1])],100,0.001)
y = zeros(len(sol4[0]))
v = zeros(len(sol4[0]))
for i in range(len(sol4[0])):
    y[i] = sol4[1][i][0]
    v[i] = sol4[1][i][1]
plot(sol4[0],y)
plot(sol4[0],sin(sol4[0]))
show()
```
